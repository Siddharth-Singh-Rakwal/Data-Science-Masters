Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

Ans. 

- loc is a label-based indexing method. It is used to select rows and columns of a DataFrame based on the labels or names of the rows and columns. This means that you can use the row and column labels to select data from a DataFrame using the loc method.

- iloc, on the other hand, is an integer-based indexing method. It is used to select rows and columns of a DataFrame based on their integer positions. This means that you can use the integer position of the rows and columns to select data from a DataFrame using the iloc method.

Q6. What do you understand about the windows function in pandas and list the types of windows
functions?

Ans. 

In Pandas, a window function, also known as a rolling or moving function, is a way to perform calculations on a sliding window of data. The window function allows you to apply a function to a subset of the data, rather than the entire dataset.

There are various types of window functions available in Pandas. Here are some of the most commonly used types:

1. Rolling window functions: These functions create a rolling window over a set number of rows and then apply a function to the data within that window. Examples of rolling window functions include rolling(), expanding(), ewm(), etc.
2. Expanding window functions: These functions apply a function to all the previous rows of the data, including the current row. Examples of expanding window functions include cumsum(), cumprod(), cummax(), and cummin().
3. Rolling window aggregation functions: These functions aggregate the data within a rolling window and return a single value for each window. Examples of rolling window aggregation functions include mean(), sum(), std(), var(), median(), quantile(), corr(), cov(), etc.
4. Time-based window functions: These functions allow you to create rolling windows based on time periods, rather than a fixed number of rows. Examples of time-based window functions include resample(), asfreq(), etc.
5. Custom window functions: These functions allow you to apply your own custom function to a rolling window of data. You can create custom window functions using the apply() method.

Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. The
program should prompt the user to enter the dates and display the result.

```python
import pandas as pd

date1 = input("Enter the first date (YYYY-MM-DD): ")
date2 = input("Enter the second date (YYYY-MM-DD): ")

date1 = pd.to_datetime(date1)
date2 = pd.to_datetime(date2)

diff = date2 - date1

days = diff.days
hours = diff.seconds // 3600
minutes = (diff.seconds % 3600) // 60

print("The difference between", date1.date(), "and", date2.date(), "is:")
print(days, "days", hours, "hours", minutes, "minutes")
```

Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.

Ans

```python
import pandas as pd

# Prompt the user to enter the file path, column name, and category order
file_path = input("Enter the file path: ")
column_name = input("Enter the column name to convert: ")
category_order = input("Enter the category order (comma-separated): ").split(",")

# Read the CSV file into a pandas DataFrame
data = pd.read_csv(file_path)

# Convert the specified column to a categorical data type
data[column_name] = pd.Categorical(data[column_name], categories=category_order, ordered=True)

# Sort the data based on the specified category order
data = data.sort_values(by=column_name)

# Display the sorted data
print(data)
```

Q10. Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.

Ans.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Prompt the user to enter the file path
file_path = input("Enter the file path: ")

# Read the CSV file into a pandas DataFrame
data = pd.read_csv(file_path)

# Convert the 'Date' column to a datetime data type
data['Date'] = pd.to_datetime(data['Date'])

# Create a pivot table to aggregate the sales data by product category and date
sales_by_category = pd.pivot_table(data, values='Sales', index='Date', columns='Category', aggfunc=sum, fill_value=0)

# Create a stacked bar chart to visualize the sales data
sales_by_category.plot(kind='bar', stacked=True, figsize=(10, 6))

# Set the chart title and axis labels
plt.title('Sales by Product Category')
plt.xlabel('Date')
plt.ylabel('Sales')

# Display the chart
plt.show()
```
