Q2. Why is request used in Flask?

Ans. request is a built-in module in Flask that provides a way to handle incoming HTTP requests in your web application. Flask uses the request module to get information about the incoming request, such as the data submitted in a form, the parameters in a query string, or the headers in the HTTP request.

Here are some common use cases for the request module in Flask:
1. Retrieving form data: When a user submits a form in a web application, the data is sent to the server as part of the HTTP request. The request module in Flask allows you to retrieve this data, which can be used to perform actions such as creating a new user account or updating a record in a database.
2. Handling query strings: When a user navigates to a URL with parameters in the query string, such as example.com/search?q=python, the request module in Flask allows you to retrieve these parameters, which can be used to perform a search or filter the data displayed in the application.
3. Accessing headers: The request module in Flask allows you to access the headers in the HTTP request, which can be used to get information about the user agent, the content type, or the authentication credentials.

Q3. Why is redirect() used in Flask?

Ans. The redirect() function is a built-in function in Flask that allows you to redirect the user to a different URL in your web application. This function is used when you want to redirect the user to a different page after they have completed an action, such as submitting a form or logging in.

Here are some common use cases for the redirect() function in Flask:
1. After submitting a form: When a user submits a form, the server typically processes the data and sends a response back to the browser. If you want to redirect the user to a different page after the form has been submitted, you can use the redirect() function. For example, you might redirect the user to a "thank you" page after they have submitted a contact form.
2. After logging in: When a user logs in to your web application, you might want to redirect them to a different page based on their role or preferences. For example, you might redirect an administrator to the admin dashboard, or a regular user to their account page.
3. Handling errors: If an error occurs in your application, you might want to redirect the user to an error page that provides more information about the error and how to fix it.

Q4. What are templates in Flask? Why is the render_template() function used?

Ans. In Flask, a template is a file that contains HTML and placeholders for dynamic content. Templates are used to create the layout and structure of web pages in a Flask application, and they allow you to separate the presentation layer from the business logic layer.

The render_template() function in Flask is used to render a template and generate the HTML that is sent to the client's web browser. This function takes the name of the template file and any variables that need to be passed to the template.

```python
# HTML Code:

<!DOCTYPE html>
<html>
  <head>
    <title>PWSkills</title>
  </head>
  <body>
    <h1>Hello, {{ name }}!</h1>
  </body>
</html>
```

```python
from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def index():
    name = 'Siddharth'
    return render_template('index.html', name=name)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
