Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans. Lookup Error acts as a base class for the exceptions that occur when a key or index used on a mapping or sequence of a list/dictionary is invalid or does not exists.

```python
# KeyError:

try:  
    a = {1:'a', 2:'b', 3:'c'}  
    print (a[4])  
except KeyError:  
    print ("Key Error Exception Raised.")
else:  
    print ("Success, no error!")
```

```python
# IndexError:

try:  
    a = ['a', 'b', 'c']  
    print (a[4])  
except IndexError:  
    print ("Index Error Exception Raised, list index out of range")
else:  
    print ("Success, no error!")
```

Q5. Explain ImportError. What is ModuleNotFoundError?

Ans. ImportError is raised when a module, or member of a module, cannot be imported. There are a two conditions where an ImportError might be raised.

1. If a module does not exist.
2. If from X import Y is used and Y cannot be found inside the module X, an ImportError is raised.

```python
# Example:

try :
    import sidd
except ImportError as e : 
    print(e)
```

ModuleNotFoundError occurs when you're trying to access or use a module that cannot be found.

```python
import sidd as sd
```
